```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random as rand
from numpy.linalg import inv
%matplotlib inline
```

```python
X = np.matrix('1 0; 1 2; 1 3; 1 4')
y = np.matrix('1; 3; 6; 8')
theta = inv(X.T.dot(X)).dot(X.T).dot(y) 
print(X.dot(theta))
theta
```

```python
plt.figure()
points = np.matrix('0; 2; 3; 4')
plt.scatter(np.matrix('0; 2; 3; 4'), y)
plt.plot([0,2,3,4], X.dot(theta))
plt.savefig('optimal.png')
```

```python
# 100 vaues between 0 and 4
random_x = np.random.random([100]) * 4
val_noise = []
for x in random_x:
    y_val = 0.51428571 + 1.77142857*x
    noise = np.random.normal(loc=0.0, scale=1.0, size=1)
    val_noise += [y_val + noise]
print(val_noise)
colors = np.random.rand(100)
plt.scatter(random_x, val_noise, c= colors)
```

```python
x_vals = np.asarray([[i] for i in random_x])
val_noise = np.asarray(val_noise)
normal_again = inv(x_vals.T.dot(x_vals)).dot(x_vals.T).dot(val_noise) 
plt.plot(x_vals, x_vals.dot(normal_again), 'r--')
```
