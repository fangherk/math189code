We are working with MNIST data, a toy dataset to test algorithms on.

```python
# let's first start w. logistic regression
# read in the data
train = np.asarray(pd.read_csv('mnist_train.csv', header = None))
test = np.asarray(pd.read_csv('mnist_test.csv', header = None))
```

```python
def predict_knn(realX, X, y, k=5):
    predY = np.zeros(realX.shape[0])
    for i in range(realX.shape[0]):
        sample = realX[i]
        # obtains the the indices are the largest k numbers 
        # basically subtract every single value in the 784 columns by the sample column
        # divide 1 over to obtain the closest k neighbors, then find their indices
        # from y, get thoce indices and the corresponding labels then predict based off
        # of the most labels in the bin
        ind = np.argpartition(1/np.linalg.norm(X - sample[:,np.newaxis].T , axis =1),
                              -k)[-k:]
        useful_list = []
        for y in yTrain[ind]:
            useful_list.append(y)
        predY[i] = np.argmax(np.bincount(useful_list)) # since the object is too deep
    return predY
```
