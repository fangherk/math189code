```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pprint
import seaborn as sns
%matplotlib inline
```

```python
csv_file = "online_news_popularity.csv"
raw_data = pd.read_csv(csv_file)
```

We will use the solution given by
$$
    (X^TX + \lambda I) \beta = X^Ty 
$$
$$
    \beta = (X^TX + \lambda I)^{-1}X^Ty.
$$
Also, note that 
$$
    y = \beta x
$$

```python
#drop the url because it is only a label. Notice that we are
# predicting shares, so we drop that too
X = raw_data.drop([" shares", "url"], 1)

# insert the bias term as the first column
lambda_bias = np.ones(X.shape[0])
X.insert(0, " bias", lambda_bias)
## Find the log value of shares
y = raw_data[" shares"]
log_y = np.log(y)
```

```python
# Create test and validation sets
# 2/3 training
# 1/3 test ---
#     (1/2) validation set
#     (1/2) test set
trainSplit = int(2*X.shape[0]/3)
testSplit = int(1*X.shape[0]/6)

train_X = X[:trainSplit]
train_y = y[:trainSplit]
train_log_y = log_y[:trainSplit]

valid_X = X[trainSplit: trainSplit+testSplit]
valid_y = y[trainSplit: trainSplit+testSplit]
valid_log_y = log_y[trainSplit:trainSplit+testSplit]

test_X = X[trainSplit+testSplit:]
test_y = y[trainSplit+testSplit:]
test_log_y = log_y[trainSplit+testSplit:]
```

$$
    \beta = (X^TX + \alpha I)^{-1}X^Ty.
$$

```python
I = np.identity(train_X.shape[1])
I[0,0]=0
lambda_RMSE = []
beta_vals = []

for i in [i for i in np.linspace(0,150,3*150)]:
    alpha = i
    beta = np.linalg.inv(train_X.T.dot(train_X) + alpha * I)
    beta = beta.dot(train_X.T).dot(train_log_y)
    
    beta_vals.append(np.sqrt(beta.dot(beta)))
    predicted= valid_X.dot(beta)
    true_val = valid_log_y
    mse = np.square(np.subtract(true_val, predicted))
    rmse = np.sqrt(np.sum(mse)/ valid_X.shape[0])
    
    lambda_RMSE.append((alpha, rmse))
```

```python
alpha_vals = []
rmse_vals = []
for alpha, rmse in lambda_RMSE:
    alpha_vals.append(alpha)
    rmse_vals.append(rmse)

min_val = min(rmse_vals)
ind = rmse_vals.index(min(rmse_vals))
print("lambda =", alpha_vals[ind])
print("value =", min_val)
```
