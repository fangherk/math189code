```python
red_wine = pd.read_csv("winequality-red.csv", sep =";")
white_wine = pd.read_csv("winequality-white.csv", sep =";")
```

```python
def eMalgo(A, n, theta, cost, likelihood, m_step, max_iter=50):
    """ 
    A = data
    theta = weights
    n = number of distributions
    likelihood = input distribution
    cost = output likelihood
    m_step = maximization step
    max_iter = choose a number of times to iterate
    """
    
    ## Setup the shapes of our values 
    r =  np.ones((X.shape[0],n)) / n
    pi = np.ones(n)/ n
    costs = [cost(X, r, pi, theta)]
    
    i =0
    while i < max_iter:
        ## E-Step calculate likelihood and divide over total sum
        r = likelihood(X, theta) * pi
        r = r / np.sum(r, axis=1)[:, np.newaxis]
        
        ## M-Step maximize with respect to data and responsibility.
        pi, theta = m_step(A, r)
        
        costs.append(cost(A, r, pi, theta))
        i += 1
    return (costs, r, pi, theta)
```

```python
def gmm(A, n, pr_alpha, max_iter=100):
    """
    A = data
    n = number of distributions
    pr_alpha = prior for dirichlet
    max_iter = number of iterations to loop through
    """
    
    # generate the S_0 as the pooled variance in dimension j
    S_0 = np.diag(np.power(np.std(A, axis=0),2)) / np.power(n,(1/X.shape[1]))
    
    # Creata named tuple for the gaussian mixture model with 2 parameters:
    # mean and covariance
    objectHolder = namedtuple("gaussian", "mean cov")
    theta = objectHolder(mean = A[np.random.randint(0, X.shape[0], n)],cov = np.tile(S_0, (n,1,1)))
    
    # Create the denominator of the MAP estimate on the mixture weights
    denominator = A.shape[0] + pr_alpha.sum() - n
    prior_nu = A.shape[1] + 2
    
    # define the maximization step with respect to our two variables mu 
    # and sigma
    def m_step(X, r):
        # get the sum of the columns
        r_sum = r.sum(axis = 0)
        # generate our MAP estimate
        pi = (r_sum + pr_alpha - 1) / denominator
        
        # generate mu maximization
        mu = ((A[:,:, np.newaxis] * r[:, np.newaxis,:]).sum(axis=0) / r_sum).T
        
        # generate sigma maximization
        sigma = np.zeros((n, A.shape[1], A.shape[1]))
        for i in range(n):
            inside = (A - mu[i])* np.sqrt(r[:,i])[:,np.newaxis]
            sigma[i] = (inside.T @ inside + S_0) / (prior_nu + r_sum[i] + A.shape[1] + 2)
        return pi, objectHolder(mean=mu, cov=sigma)
    
    ## Set up the likelihood.
    # 1. create the shape
    # 2. loop through
    def likelihood(A, theta):
        # shape num. of examples by n distributions
        ll = np.zeros((X.shape[0], n))
        # for each distribution, take the MVN w. respect to the generated theta values
        for i in range(n):
            ll[:, i] = stats.multivariate_normal.pdf(A, theta.mean[i], theta.cov[i] + 1e-4*np.eye(A.shape[1]))
        return ll
    
            
    def cost(X, r, pi, theta):
        # combine all the terms to produce the expected complete data log-likelihood
        log_prior = np.sum(np.log(stats.invwishart.pdf(theta.cov[i], df=prior_nu, scale=S_0 )) for i in range(n)) 
        + np.log(stats.dirichlet.pdf(pi, alpha=pr_alpha))
        
        ## get the distribution prior
        dist_prior = (r * np.log(pi)[np.newaxis,:]).sum()
        
        # get the likelihood term prior
        likelihood_term = r*np.log(likelihood(X, theta))
        likelihood_term = likelihood_term[r > 1e-10].sum()
        return likelihood_term + dist_prior + log_prior
    
    return eMalgo(X, n, theta, cost, likelihood, m_step, max_iter)
    
```

```python
X = np.concatenate((red_wine.as_matrix(), white_wine.as_matrix()), axis=0)
y = np.zeros((X.shape[0]))
y[:red_wine.shape[0]] = 1
X.shape
```

```python
n = 2 
obj, r, pi, theta = gmm(X, n, np.ones(2), max_iter=40)
```

```python
plt.plot(obj)
plt.title("Mixture Model")
plt.xlabel("Iter")
plt.ylabel("Expected Complete Data Log Likelihood")
```

```python
np.set_printoptions(precision=2)

plt.imshow(metrics.confusion_matrix(y, ~np.argmax(r, axis=1).astype(bool)), cmap="Greys_r")
plt.xlabel("True")
plt.ylabel("Predicted")
plt.colorbar()
ticks = np.arange(2)
colors = ["white", "red"]
plt.xticks(ticks, colors)
plt.yticks(ticks, colors)
```
